```python
def vader_analysis(compound):
    if compound > 0.7:
        return 'Positive'
    elif compound < 0.7 and compound > -0.7 :
        return 'Neutral'
    else:
        return 'Negative'
```

## For topic 0

```python
df = pd.read_csv('topic-1_market.csv') 
df['compound'] = df['text'].apply(lambda x:analyser.polarity_scores(x)['compound'])
```

```python
df['Sentiment']=df['compound'].apply(vader_analysis)
df.tail()
```

```python
df = df.drop('business_id',axis = 1)
# df = df.drop('pos',axis = 1)
df
```

```python
comp = df['compound'].mean()
comp
```

```python
final_overall_sentiment = vader_analysis(comp)
print(final_overall_sentiment)
```
